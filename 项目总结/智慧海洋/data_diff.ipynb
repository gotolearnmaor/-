```python
import pandas as pd
```

```python
train =pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
def difference(df):
    df['x_diff'] = df.groupby('ship')['x'].diff(1)
    df['y_diff'] = df.groupby('ship')['y'].diff(1)
    df['v_diff'] = df.groupby('ship')['v'].diff(1)
    df['g_diff'] = df.groupby('ship')['g'].diff(1)
    
```

```python
difference(train)
difference(test)
```

```python
train.fillna(method='bfill')
test.fillna(method='bfill')
```

```python
train.to_csv("train.csv",index = None)
test.to_csv("test.csv",index = None)
```
