```python
import tqdm
import os
import pandas as pd
import numpy as np
```

```python
# 首先获得文件路径
test_path = "hy_round1_testA_20200102"
train_path = "hy_round1_train_20200102"

#获取当前文件下的所有文件名,返回字符串
train_files = os.listdir(train_path)
test_files = os.listdir(test_path)
print(len(train_files),len(test_files))
```

```python
def feature(df):   
    n_033 = len(df[df["v"]< 0.33])
    df['speed_033']=n_033

    n_6 = len(df[df["v"]>= 6])
    df['speed_6'] = n_6


    n_23 =len(df[(df["v"]< 6) & (df["v"]>2) ])
    df['speed_23'] = n_23

    g_90 =len(df[df['g']<90])
    df["g_90"] = g_90

    g_100 =len(df[(df['g']>100) &(df['g']<300)])
    df["g_100"] = g_100

    g_300 =len(df[df['g']>300])
    df["g_300"] = g_300
```

```python
def clean(df):
    dt_index = df[df['v'] == 0].index
    dt_index = dt_index.values.tolist()
    data = dt_index.copy()
    for i in range(len(data)-1):        
        data[i] = data[i+1] - data[i]       
    res =[]
    ind=0
    count = 0
    while (ind+count) < len(data)-1 :

        while (data[ind + count] == 1) & (ind+count <len(data)-1):
            count+=1
        if count >3:
            for num in range(count):
                res.append(ind+num)
        ind = ind+count+1
        count = 0

    result =[]
    for i in res:
        result.append(dt_index[i])
    df.drop(result,inplace=True)
    return result
```

```python
ret = []
for file in tqdm.tqdm(train_files):
    df = pd.read_csv(f'{train_path}/{file}')
    clean(df)
    feature(df)
    df.to_csv(f'{train_path}/{file}',index = None)
```

```python
#tqdm 可用于显示进度条，可用于任意迭代器
#将训练集整合成一个数据集
ret = []
for file in tqdm.tqdm(train_files):
    df = pd.read_csv(f'{train_path}/{file}')  
    df.columns = ['ship','x','y','v','g','time','type','speed_033','speed_6','speed_23','g_90','g_100','g_300']
    
    
    ret.append(df)
df = pd.concat(ret)
```

```python
df.columns = ['ship','x','y','v','g','time','type','speed_033','speed_6','speed_23','g_90','g_100','g_300']
```

```python
df.to_csv("train.csv",index = None)
```
